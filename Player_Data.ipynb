```python
import pandas as pd
import numpy as np
```

```python
#Uploading the Corsica hockey html
url = 'https://www.corsicahockey.com/nhl/players/nhl-player-ratings-rankings'
```

## Extracting Data

```python
#Extracting the data from the Corsica hockey html tables
tables = pd.read_html(url)
tables
```

```python
 type(tables)
```

```python
corsica_df = tables[0]
corsica_df.head()
```

## Transforming Data

```python
#replacing the NaN  with -
corsica_df.fillna('-')
```

```python
#Drop the "Opponent" column
new_corsica_df = corsica_df.drop(columns="Opponent")
new_corsica_df
```

```python
#Drop the "Ranking" column
new2_corsica_df = new_corsica_df.drop(columns="Ranking")
new2_corsica_df
```

```python
updtd_data_df = new2_corsica_df.rename(columns= {'Team': 'franchise', 'Player': 'player_name','Age': 'age', 'Rating': 'rating',
                                                 'Position': 'position', 'GP': 'games_played', 'Shoots': 'shoots_L/R'})

updtd_data_df.columns
```

```python
player_data_df = updtd_data_df[['franchise', 'player_name', 'age', 'position', 'rating', 'games_played', 'shoots_L/R']]
player_data_df
```
